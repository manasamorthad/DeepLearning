```python
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical

train = pd.read_csv('/content/drive/MyDrive/DL_datasets/sign_mnist_train.csv')
test  = pd.read_csv('/content/drive/MyDrive/DL_datasets/sign_mnist_test.csv')

X_train = train.drop('label', axis=1).values / 255.0
X_test  = test.drop('label', axis=1).values / 255.0

y_train_raw = train['label'].values
y_test_raw  = test['label'].values

# remap labels to continuous values
unique_labels = sorted(np.unique(y_train_raw))
label_map = {label: idx for idx, label in enumerate(unique_labels)}

y_train_mapped = np.array([label_map[label] for label in y_train_raw])
y_test_mapped  = np.array([label_map[label] for label in y_test_raw])

num_classes = len(unique_labels)

y_train = to_categorical(y_train_mapped, num_classes)
y_test  = to_categorical(y_test_mapped, num_classes)

print(X_train.shape, y_train.shape)
print("Classes:", num_classes)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

def create_model(num_classes):
    model = Sequential([
        Input(shape=(784,)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    return model
```

```python
from tensorflow.keras.optimizers import SGD

model_bgd = create_model(num_classes)

model_bgd.compile(
    optimizer=SGD(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_bgd = model_bgd.fit(
    X_train, y_train,
    epochs=80,
    batch_size=len(X_train),
    validation_data=(X_test, y_test),
    verbose=1
)

loss, acc = model_bgd.evaluate(X_test, y_test, verbose=0)

print("Batch Gradient Descent Results")
print("Test Loss:", loss)
print("Test Accuracy:", acc)
```

```python
model_sgd = create_model(num_classes)

model_sgd.compile(
    optimizer=SGD(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_sgd = model_sgd.fit(
    X_train, y_train,
    epochs=15,
    batch_size=1,
    validation_data=(X_test, y_test),
    verbose=1
)

loss, acc = model_sgd.evaluate(X_test, y_test, verbose=0)

print("Stochastic Gradient Descent Results")
print("Test Loss:", loss)
print("Test Accuracy:", acc)
```

```python
from tensorflow.keras.optimizers import SGD

model_minibatch = create_model(num_classes)

model_minibatch.compile(
    optimizer=SGD(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_minibatch = model_minibatch.fit(
    X_train, y_train,
    epochs=15,
    batch_size=32,   # рџ”Ґ Mini-batch
    validation_data=(X_test, y_test),
    verbose=1
)

loss_mb, acc_mb = model_minibatch.evaluate(X_test, y_test, verbose=0)

print("\nMini-Batch GD Results")
print("Test Loss:", loss_mb)
print("Test Accuracy:", acc_mb)
```

```python
from tensorflow.keras.optimizers import SGD

# Momentum SGD Model
model_momentum = create_model(num_classes)

model_momentum.compile(
    optimizer=SGD(
        learning_rate=0.01,
        momentum=0.9   #
    ),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_momentum = model_momentum.fit(
    X_train, y_train,
    epochs=15,
    batch_size=32,   # usually mini-batch is used with momentum
    validation_data=(X_test, y_test),
    verbose=1
)

loss_m, acc_m = model_momentum.evaluate(X_test, y_test, verbose=0)

print("\nSGD with Momentum Results")
print("Test Loss:", loss_m)
print("Test Accuracy:", acc_m)
```

```python
# Nesterov SGD Model
model_nesterov = create_model(num_classes)

model_nesterov.compile(
    optimizer=SGD(
        learning_rate=0.01,
        momentum=0.9,
        nesterov=True
    ),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_nesterov = model_nesterov.fit(
    X_train, y_train,
    epochs=15,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

loss_n, acc_n = model_nesterov.evaluate(X_test, y_test, verbose=0)

print("\nSGD with Nesterov Results")
print("Test Loss:", loss_n)
print("Test Accuracy:", acc_n)
```

```python
from tensorflow.keras.optimizers import Adagrad

model_adagrad = create_model(num_classes)

model_adagrad.compile(
    optimizer=Adagrad(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_adagrad = model_adagrad.fit(
    X_train, y_train,
    epochs=15,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

loss_ag, acc_ag = model_adagrad.evaluate(X_test, y_test, verbose=0)

print("\nAdagrad Results")
print("Test Loss:", loss_ag)
print("Test Accuracy:", acc_ag)
```

```python
from tensorflow.keras.optimizers import RMSprop

model_rmsprop = create_model(num_classes)

model_rmsprop.compile(
    optimizer=RMSprop(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_rmsprop = model_rmsprop.fit(
    X_train, y_train,
    epochs=15,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

loss_rms, acc_rms = model_rmsprop.evaluate(X_test, y_test, verbose=0)

print("\nRMSProp Results")
print("Test Loss:", loss_rms)
print("Test Accuracy:", acc_rms)
```
